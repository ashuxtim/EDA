```python
con1=visa_df['Continent']=='Asia'
con2=visa_df['Case_Status']=='Granted'
con3=visa_df['Case_Status']=='Denied'
grant_con=con1&con2
denied_con=con1&con3

l1=len(visa_df[grant_con])
l2=len(visa_df[denied_con])

print(f"{l1} people were granted visa from Asia")
print(f"{l2} people were denied visa from Asia")
```

```python
labels=visa_df['Continent'].value_counts().keys()
for i in labels:
    con1=visa_df['Continent']==i
    con2=visa_df['Case_Status']=='Granted'
    con3=visa_df['Case_Status']=='Denied'
    grant_con=con1&con2
    denied_con=con1&con3
    
    l1=len(visa_df[grant_con])
    l2=len(visa_df[denied_con])
    
    print(f"{l1} people were granted visa from {i}")
    print(f"{l2} people were denied visa from {i}")
```

```python
labels=visa_df['Continent'].value_counts().keys()
cert=[]
deni=[]

for i in labels:
    con1=visa_df['Continent']==i
    con2=visa_df['Case_Status']=='Granted'
    con3=visa_df['Case_Status']=='Denied'
    grant_con=con1&con2
    denied_con=con1&con3
    
    l1=len(visa_df[grant_con])
    l2=len(visa_df[denied_con])
    
    cert.append(l1)
    deni.append(l2)

d1=pd.DataFrame(zip(labels,cert,deni),columns=["Continent","Granted","Denied"])
d1.set_index('Continent')
```

$pd.crosstab()$

```python
col1=visa_df['Continent']
col2=visa_df['Case_Status']
result1=pd.crosstab(col1,col2)
result1
```

```python
col1=visa_df['Continent']
col2=[visa_df['Case_Status'],visa_df['Education_Level']]
result2=pd.crosstab(col1,col2)
result2

# col1 is generally index
# col2 is columns
```

```python
col1=visa_df['Education_Level']
col2=[visa_df['Case_Status'],visa_df['Continent']]
result3=pd.crosstab(col1,col2)
result3.plot(kind='bar')
plt.show()
```

**Numerical vs Numerical**

$Scatter$ $Plots$

- plt.scatter

- will take two arguments x-axis and y-axis

- Both variables should be numerical

- It provides relationship between two variables

    - Positively related

    - Negatively related

    - No relation

```python
x=[1,2,3,4,5]
y=[11,12,13,14,15]

# (1,11)  (2,12)  (3,13)  (4,14)  (5,15)
plt.scatter(x,y)
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,y)
```

```python
x=[i for i in range(-4,5)]
y=[i*i for i in range(-4,5)]
plt.scatter(x,y)
plt.plot(x,y)

# y=x^2
```

```python
col1=visa_df['Prevailing_Wage']
col2=visa_df['No_of_Employees']

plt.scatter(col1,col2)
plt.plot(col1,col2)
```

**Correlation-Coefiecient**

- Denoted with r

- Pearson correlation coefficient

- r varies from -1 to 1

- r = -1 to 0 : Negatively coorelation

- r = 0 to 1 : Positively coorelation

- r = 0 : No relation

  

```python
visa_df.corr(numeric_only=True)
```

$Heatmap$

```python
# sns.heatmap(<covv matrix>)
# heatmap available from seaborn package
# Line your right side min=0 max=1
# Entire matrix value ranges from 0 to 1

corr=visa_df.corr(numeric_only=True)

sns.heatmap(corr,annot=True)
```
