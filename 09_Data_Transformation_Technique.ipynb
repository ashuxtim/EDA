**Data Transformation Techniques**

- Generally used for to convert Normal distribution
- Because all statistical math analysis by assumption Data follows Normal distribution
- It also avoid skew-ness
- We have some important transfomations:
    - log transformation
    - exponential transformation
    - Reciprocal transormation
    - Square root transformation
    - Power transformation 

$Step-1$

**Read the required packages**

```python
import numpy as np
import matplotlib.pyplot as plt
```

$Step-2$

**Read the data**

$Exponential-data$

```python
exp_data=np.random.exponential(size=10000)
exp_data[:10]

# We are consider exponential data set with random 1000 observations
# Using numpy package
```

```python
plt.hist(exp_data,bins=30,label='Exponential')
plt.legend()
plt.show()
```

$Normal-data$

```python
norm_data=np.random.normal(size=1000)
plt.hist(norm_data,bins=30,label='Normal')
plt.legend()
plt.show()
```

$Step-3$

**Log Transformation**

- np.log is used for log transformation
- Generally log tranformation will not convert data into normal
- It avoids skew-ness
- np.log means natural logarithm base = e

```python
log_data=np.log(exp_data)
log_data[:10]
```

```python
exp_data[:10]
```

```python
plt.figure(figsize=(20,7))
plt.subplot(1,2,1).hist(exp_data,bins=30,label='Exponential Data',color='green')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=30,label='Log Transformation',color='violet')
plt.legend()
plt.show()
```

$Step-4$

**Reciprocal transformation**

```python
rec_data=1/exp_data
rec_data[:10],exp_data[:10]
```

```python
plt.figure(figsize=(20,7))
plt.subplot(1,2,1).hist(exp_data,bins=30,label='Exponential Data',color='green')
plt.legend()
plt.subplot(1,2,2).hist(rec_data,bins=30,label='Reciprocal Transformation',color='brown')
plt.legend()
plt.show()
```

- Reciprocal Transformation fails when data has zero values

$Step-5$

**Square root Transformation**

```python
sroot_data=np.sqrt(exp_data)
sroot_data[:10],exp_data[:10]
```

```python
plt.figure(figsize=(20,7))
plt.subplot(1,2,1).hist(exp_data,bins=30,label='Exponential Data',color='green')
plt.legend()
plt.subplot(1,2,2).hist(sroot_data,bins=30,label='Square Root Transformation',color='brown')
plt.legend()
plt.show()
```

$Step-6$

**Power Tranformation**

- It is related to sklearn package
- Package name - sklearn.preprocessing
- Method name - Power Transformer
- Inside Box-Cox, yeo-jhonson

```python
exp_data=np.random.exponential(size=10000)

from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()
pt_data=pt.fit([exp_data]).transform([exp_data])
pt_data
```

```python
plt.hist(pt_data[0])
plt.show()
```
