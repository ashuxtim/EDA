$map$

- First consider one categorical column
- Observe how many unique labels are there
- Assume here we are taking case_status column
- case_status has two unique labels
    - Denied
    - Granted
- Create a dictionary by providing key:value
- Here the key = Denied and the value = 1
- key = Granted and the value = 0
- dictionary will look like : {'Granted':0,'Denied':1}

```python
visa_df['Case_Status'].unique()
```

```python
d1={'Granted':0,'Denied':1}
visa_df['Case_Status'].map(d1)

# After mapping, in which column you want to apply the data
# You want to create a new column
# You want to use old column

# visa_df['Case_Status1']=visa_df['Case_Status'].map(d1)   ---> New column

visa_df['Case_Status']=visa_df['Case_Status'].map(d1)      # ---> Overwrite (Recommended)
```

```python
visa_df.head()
```

```python
d1={}
labels=visa_df['Continent'].unique()
for i in range(len(labels)):
    d1[labels[i]]=i

visa_df['Continent']=visa_df['Continent'].map(d1)
visa_df.head()
```

```python
file_loc="C:\\Users\\ashux\\OneDrive\\Python Notes\\New folder\\visa_application_data.csv"
visa_df=pd.read_csv(file_loc)
visa_df.head()
```

- np.where is a kind of if-else
- np.where is useful for binary condition
- Which means it is useful for the columns having two labels
- np.where will take three parameters
    - condition
    - True value
    - False value
- I want to replace case_status column with 1 when it is Denied
- Otherwise it is zero
- Con = visa_df['Case_Status'] == "Denied"
- True value = 1  False value = 0
- np.where(con, True value, False value)

```python
con=visa_df['Case_Status']=='Denied'
visa_df['Case_Status']=np.where(con,1,0)
visa_df
```

**Method-3**

$one - hot - encoder$

pd.getdummies()

- one hot encoder means at a time only one will be ON, other will be OFF
- ON represents with 1 and OFF represents with 0
- One hot encoder creates new extra columns based on labels in the columns
- For example Case_Status has two unique labels
- If you apply one hot encoder on Case_Status it creates two extra columns
    - Case_Status_Denied
    - Case_Status_Granted


|Case Status|Case_Status_Denied|Case_Status_Granted|
|-----------|------------------|-------------------|
|Denied     |        1         |        0          |
|Granted    |        0         |        1          |

**Advantages**

- One hot encoder gives independancy between the variables
- It is one of the important property before apply ML models
- Variables are independent to each other
- Variables are perpendicular to each other
- Variables are Orthogonal to each other
- Variables are 90 degrees phase shift

**Disadvantages**

- Assume that a column has 100 unique labels
- If we apply one hot encoder it will create 100 new columns
- Your data matrix become sparse
- We require more time and more memory to compute the calculations
- This is called **Curse of Dimensionality**

```python
# Read the data
file_loc="C:\\Users\\ashux\\OneDrive\\Python Notes\\New folder\\visa_application_data.csv"
visa_df=pd.read_csv(file_loc)
visa_df.head()

# Drop the ID before applying one hot encoder
visa_df.drop(['Case_ID','Applicant_Name'],axis=1,inplace=True)

# Apply one hot encoder
pd.get_dummies(visa_df,dtype='int')
```

```python
# Read the data
file_loc="C:\\Users\\ashux\\OneDrive\\Python Notes\\New folder\\visa_application_data.csv"
visa_df=pd.read_csv(file_loc)

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

# We cant directly pass the data
# Apply the loop

cat_cols=visa_df.select_dtypes(include='object').columns

for i in cat_cols:
    visa_df[i]=le.fit_transform(visa_df[i])

visa_df.head()
```
